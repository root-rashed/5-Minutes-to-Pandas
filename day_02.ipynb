Dictionary to DataFrame or Series

```python
import pandas as pd
```

```python
data = {
    'name':['rashed','sabit','prio'],
    'age':[23,24,23],
    'city':['jamalpur','serpur','rangpur']
}
```

```python
data
```

```python
name_series = pd.Series(data['name'],name='Name')
name_series
```

```python
df = pd.DataFrame(data)
df
```

```python
age_series = df['age']
age_series
```

```python
city_series = df['city']
city_series
```
